```python
def add_folder_structure(driver, repo_name, folder_structure):
    with driver.session() as session:
        # Create or find the repository node
        repo_node = session.run(
            "MERGE (repo:Repository {name: $repo_name}) RETURN repo",
            repo_name=repo_name
        ).single()["repo"]

        # Traverse the folder structure and create nodes and relationships
        current_node = repo_node
        for folder in folder_structure:
            # Create or find the folder/file node
            folder_node = session.run(
                "MERGE (folder:Folder {name: $folder}) RETURN folder",
                folder=folder
            ).single()["folder"]

            # Create the CONTAINS relationship between parent and folder
            session.run(
                "MATCH (parent), (child) "
                "WHERE id(parent) = $parent_id AND id(child) = $child_id "
                "MERGE (parent)-[:CONTAINS]->(child)",
                parent_id=current_node.id, child_id=folder_node.id
            )
            
            # Move to the next folder level
            current_node = folder_node
```
